```python
import numpy as np
import pandas as pd
import yfinance
import random
```

```python
url = "https://raw.githubusercontent.com/Ritesh-k99/Portfolio-Creation-Optimization/main/list28.csv"
tickers = pd.read_csv(url)
ticker = yfinance.Ticker("^NSEI")
days = ticker.history(interval ="1d",start = "2018-01-01",end = "2019-12-31").shape[0]
#shape[0] gives the count of no of trading days in between the two dates.
```

```python
total = None
for symbol in tickers['Symbol']:
  
  ticker = yfinance.Ticker(symbol)
  close = ticker.history(interval="1d",start="2018-01-01",end="2020-09-24")[['Close']].pct_change(1)
  close.rename(columns={'Close':symbol},inplace=True)

  if close.shape[0] < days:
    continue
    
  if total is None:
    total = close
  else:
    total = total.merge(close,left_index = True, right_index = True,suffixes = ("",""))

  print(symbol,total.shape)
```

```python
total
```

```python
corr_table = total.corr()
corr_table['stock1'] = corr_table.index #row axis pe change karrha ha
corr_table = corr_table.melt(id_vars = 'stock1',var_name = 'stock2').reset_index(drop = True)
```

```python
corr_table = corr_table[corr_table['stock1'] < corr_table['stock2']].dropna()
corr_table['abs_value'] = np.abs(corr_table['value'])
highest_corr = corr_table.sort_values("abs_value", ascending = True)
highest_corr.head(5)
```

```python
l1 = list(highest_corr['stock1'].head(5))
l2 = list(highest_corr['stock2'].head(5))
l3 = l1 + l2
rs = random.sample(set(l3),3)
rs
```
